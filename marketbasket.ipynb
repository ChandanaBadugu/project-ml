#  Modeling

```python
frequent_itemsets = apriori(df,
                            min_support = .006,
                            max_len = 3,
                            use_colnames = True)
```

```python
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets,
                            metric = 'support',
                            min_threshold=0.1)
```

```python
rules
```

```python
filtered_rules = rules[(rules['antecedent support'] > 0.02)&
                        (rules['consequent support'] >0.01) &
                        (rules['confidence'] > 0.2) &
                        (rules['lift'] > 1.0)]
```

```python
rules2 = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules2 = rules2.sort_values(['lift'], ascending=False)
rules2.head()
```
